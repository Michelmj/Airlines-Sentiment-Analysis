Now that we have an idea of what is going on, let us look at each sentiment for each airline and see what the main words are

We can do a sum of the number of tweets and add some plots for each airline as it relates to the sentiments; positive, negative, and neutral. Plotting always helps

```python
sentair.groupby('airline').head()
```

```python
print('Total number of sentiments for tweets for the airline \n', sentair.groupby('airline')['airline_sentiment'].count().sort_values(ascending = False))
airline = ['US Airways', 'United', 'American', 'Southwest', 'Delta', 'Virgin America']
plt.figure(1, figsize = (60, 30))
for m in airline: 
    indices = airline.index(m)
    plt.subplot(2, 3, indices + 1)
    sentair0 = sentair[sentair['airline'] == m]
    count = sentair0['airline_sentiment'].value_counts()
    Index = [1,2,3]
    plt.bar(Index, count, color = ['yellow', 'red', 'blue'])
    plt.xticks(Index, ['negative', 'neutral', 'positive'])
    plt.xlabel('Sentiments', fontsize = 55)
    plt.ylabel('Count', fontsize = 55)
    plt.title('Count of Different Sentiments by ' + m, fontsize = 53)
    
```

It looks as if Virgin America is doing a lot better at keeping customers happy. It is interesting how most people have negative sentiments about these six airline industries. What are the most used words in Positive and Negative tweets?
There is something called "WordCloud" in python that can help us
find all the words we need. Since Negative words are the most 
frequent ones, we are going to take a look at them first. What we are going to look for as the most frequent words, are going to be in bigger letters; and as we see words with smaller letters, we will understand they are less frequently used

```python
from wordcloud import WordCloud, STOPWORDS
```

Wordcloud is used for word visualization techniques. We will take a look at the negative words first

```python
newsentair = sentair[sentair['airline_sentiment'] == 'negative']
words = ' '.join(newsentair['text'])
worD = " ".join([word for word in words.split()
                     if 'http' not in word and 
                      not word.startswith('@')
                     and word != 'RT'])
wordcloud = WordCloud(stopwords = STOPWORDS, 
                     background_color = 'yellow',
                     width = 3500, 
                     height = 3000,
                     ).generate(worD)
plt.figure(1, figsize = (20, 18))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

Now let us look at the positive sentiments given

```python
newsentair = sentair[sentair['airline_sentiment'] == 'positive']
words = ' '.join(newsentair['text'])
Word = " ".join([word for word in words.split()
                   if 'http' not in word and not 
                    word.startswith('@') and word != 'RT'])
wordcloud = WordCloud(stopwords = STOPWORDS, 
                     background_color = 'blue',
                     width = 3500, 
                    height = 3000).generate(Word)
plt.figure(1, figsize = (20, 18))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
newsentair = sentair[sentair['airline_sentiment'] == 'neutral']
words = ' '.join(newsentair['text'])
word = " ".join([word for word in words.split()
                   if 'http' not in word and not 
                   word.startswith ('@') and word != 'RT'])
wordcloud = WordCloud(stopwords = STOPWORDS, 
                     background_color = 'red', 
                     width = 3500,
                     height = 3000).generate(word)
plt.figure(1, figsize = (20, 18))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

We can see there is somewhat a difference between the positive, negative, and neutral sentiment portayed here. But, some words may appear in both neutral and negative, or neutral and positive because each person views certain things differently. Now it would be good to find out how many times some of these words are used.

We are going to see the highest frequency words in positive, negative and neutral sentiments

```python
def freq(str):
    str = str.split()
    str1 = []
    
    for i in str:
        if i not in str1:
            str1.append(i)
            
    for i in range(0, len(str1)):
        if(str.count(str1[i]) > 50):
            print('Frequence of', str1[i], 'is:', 
                  str.count(str1[i]))
print(freq(word))
```
